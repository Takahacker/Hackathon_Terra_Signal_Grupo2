```python
history_sdf = (
    spark.read
         .option("header", True)
         .option("inferSchema", True)
         .csv(file_path)
)

history_sdf.printSchema()
history_sdf.show(5)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.churn")
```

```python
(
    history_sdf
    .write
    .mode("overwrite")  # cuidado: sobrescreve se já existir
    .saveAsTable("workspace.churn.history")
)
```

```python
df = spark.table("workspace.churn.churn_reason_final")

df.coalesce(1) \
  .write.mode("overwrite") \
  .option("header", "true") \
  .csv("dbfs:/FileStore/tables/churn_reason_final_csv")
```
